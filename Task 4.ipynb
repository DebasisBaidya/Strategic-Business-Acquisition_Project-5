**Task 4.1 - Assign Scores**

````python
import pandas as pd

# Assuming `df` contains engagement-related data
df_engagement = engagement[[
    'MSISDN/Number',
    'session_frequency',
    'session_duration',
    'total_traffic','cluster'
]].copy()

# Assuming `aggregated` contains experience-related data
df_experience = aggregated[[
    'MSISDN/Number',
    'avg_tcp_retransmission',
    'avg_rtt',
    'avg_throughput', 'experience_cluster'
]].copy()

# Show all columns
pd.set_option('display.width', 1000)

# Display the extracted DataFrames
print("Engagement DataFrame:")
print(df_engagement.head())

print("\nExperience DataFrame:")
print(df_experience.head())
````

````python
import numpy as np
import pandas as pd

def assign_scores(df_engagement, df_experience):
    """Assigns engagement and experience scores."""

    # Get the cluster ID of the cluster with the min average session duration and min total traffic
    min_engagement_cluster = df_engagement.groupby('cluster')[['session_duration', 'total_traffic']].mean().sum(axis=1).idxmin()

    # Get the cluster ID of the cluster with the highest average TCP retransmission and highest average RTT
    max_experience_cluster = df_experience.groupby('experience_cluster')[['avg_tcp_retransmission', 'avg_rtt']].mean().sum(axis=1).idxmax()

    # Calculate cluster centers once
    engagement_centers = df_engagement.groupby('cluster')[['session_frequency', 'session_duration', 'total_traffic']].mean()
    experience_centers = df_experience.groupby('experience_cluster')[['avg_tcp_retransmission', 'avg_rtt', 'avg_throughput']].mean()

    # Assign engagement scores
    def assign_engagement_score(row):
        return np.linalg.norm(row[['session_frequency', 'session_duration', 'total_traffic']].values - engagement_centers.loc[min_engagement_cluster].values)

    df_engagement['engagement_score'] = df_engagement.apply(assign_engagement_score, axis=1)

    # Assign experience scores
    def assign_experience_score(row):
        return np.linalg.norm(row[['avg_tcp_retransmission', 'avg_rtt', 'avg_throughput']].values - experience_centers.loc[max_experience_cluster].values)

    df_experience['experience_score'] = df_experience.apply(assign_experience_score, axis=1)

    # Merge engagement and experience score into the same data frame by MSISDN/Number
    df_engagement_scores = df_engagement[['MSISDN/Number', 'engagement_score']].copy()
    df_experience_scores = df_experience[['MSISDN/Number', 'experience_score']].copy()

    scores_df = pd.merge(df_engagement_scores, df_experience_scores, on='MSISDN/Number', how='inner')
    return scores_df

# Call the function
try:
    scores_df = assign_scores(df_engagement, df_experience)
    print(scores_df)
except Exception as e:
    print(f"An error occurred: {e}")
````

### Summary of Findings

1. **Data Preparation**:
   - **Created `df_engagement`**: Extracted relevant columns from the engagement data.
   - **Created `df_experience`**: Extracted relevant columns from the experience data.

2. **Display Settings**:
   - Set display width to 1000 for better visibility of DataFrame columns.

3. **Function Definition**:
   - **Defined `assign_scores` Function**:
     - **Identified Clusters**: Found the cluster with the lowest average session duration and traffic for engagement, and the cluster with the highest TCP retransmission and RTT for experience.
     - **Calculated Averages**: Computed average metrics for each cluster.
     - **Assigned Scores**: Used Euclidean distance to calculate engagement and experience scores.
     - **Merged Scores**: Combined scores into a single DataFrame based on `MSISDN/Number`.

4. **Function Execution**:
   - **Executed the Function**: Called `assign_scores` and printed the resulting DataFrame, handling any errors that occurred.

### Output
- The result is a DataFrame (`scores_df`) containing `MSISDN/Number`, `engagement_score`, and `experience_score`. 

### Error Handling
- Included error handling to catch and print any issues during execution.

**Task 4.2 - Satisfaction Score**

````python
import numpy as np
import pandas as pd

def calculate_satisfaction_score(df):
    """Calculates the average satisfaction score and identifies satisfied customers."""
    
    # Calculate the satisfaction score
    df['satisfaction_score'] = (df['engagement_score'] + df['experience_score']) / 2
    
    # Get the top 10 satisfied customers
    top_10_satisfied = df.nlargest(10, 'satisfaction_score')
    
    # Print the top 10 satisfied customers
    print("Top 10 Satisfied Customers:\n", top_10_satisfied[['MSISDN/Number', 'satisfaction_score']])
    
    return df

updated_df = calculate_satisfaction_score(scores_df)  # Call the function with the scores DataFrame
# Print the updated DataFrame
print("\nUpdated DataFrame:\n",updated_df.head())
````

**Task 4.4 - Clustering Satisfaction and Experience**

````python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

def cluster_satisfaction_experience(df):
    """Clusters users based on satisfaction and experience scores."""
    
    # Clustering based on k-means using engagement and experience
    scaler = StandardScaler()
    scaled_scores = scaler.fit_transform(df[['engagement_score', 'experience_score']])
    
    kmeans = KMeans(n_clusters=2, random_state=42)
    df['satisfaction_cluster'] = kmeans.fit_predict(scaled_scores)
    
    # Print the first 20 cluster values
    print(f"\nCluster values:\n {df[['MSISDN/Number', 'satisfaction_cluster']].head(20)}")
    
    return df

# Now call the clustering function
clustered_df = cluster_satisfaction_experience(updated_df)
````
